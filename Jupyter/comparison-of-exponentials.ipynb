```python
import scipy
import numpy as np
import matplotlib.pyplot as plt

# definition of the exponential
a    = -3
tmax = (- 10 / a)

# definition of the timesteps
t = np.linspace(0, tmax, round(tmax * 100))

# definition of the function
y0 = np.power(t, 0) * np.exp( a * t )
y1 = np.power(t, 1) * np.exp( a * t )
y2 = np.power(t, 2) * np.exp( a * t )
y3 = np.power(t, 3) * np.exp( a * t )

# plotting
fig = plt.figure(figsize = (10, 4))
# plt.plot(t, y0, linewidth = 1, color = 'g')
plt.plot(t, y1, linewidth = 1, color = 'r')
plt.plot(t, y2, linewidth = 1, color = 'b')
plt.plot(t, y3, linewidth = 1, color = 'k')
plt.grid()
plt.xlabel('t')
plt.ylabel('y(t)')
plt.show()
```
