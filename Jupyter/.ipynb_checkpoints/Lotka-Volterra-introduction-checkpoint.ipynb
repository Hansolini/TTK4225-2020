```python
# load the necessary packages
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# define the function that returns dz/dt
def myModel(z, t):
    #
    # parameters
    alpha = 1.1
    beta  = 0.4
    gamma = 0.4
    delta = 0.1
    #
    # individual variables
    x = z[0]
    y = z[1]
    #
    # individual derivatives
    dxdt  =   alpha * x - beta  * x * y
    dydt  = - gamma * y + delta * x * y
    #
    # whole Jacobian
    dzdt = [ dxdt, dydt ]
    return dzdt

# define the initial condition
z0 = [ 2, 5 ]

# define the time points where the solution is computed
n    = 100
tmax = 20
t    = np.linspace(0, tmax, n)

# solve the ODE
z = odeint(myModel, z0, t)

# get the individual variables
x = z[:,0]
y = z[:,1]

# plot the time evolution
plt.subplot(1,2,1)
plt.plot(t, x, label='prey', linewidth=1)
plt.plot(t, y, label='predator', linewidth=1)
plt.grid()
plt.legend(loc = 'upper right')
plt.xlabel('time evolution')

# plot the phase portrait
plt.subplot(1,2,2)
plt.plot(x, y, linewidth=1)
plt.grid()
plt.xlabel('prey')
plt.ylabel('predator')
plt.title('phase portrait')

plt.show()

```
